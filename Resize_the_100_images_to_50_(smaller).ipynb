#### "Resize the 100 images to 50% (smaller)" using OpenCV

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r"/content/cat.png", 1)
# Loading the image

half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)

Titles =["Original", "Half"]
images =[image, half]

for i in range(2):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()
```
